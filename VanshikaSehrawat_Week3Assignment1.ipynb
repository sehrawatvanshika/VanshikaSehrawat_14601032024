Q4. Distribution Analysis
Create histograms for:
1. rainfall_mm  
2. temperature_c  
3. fertilizer_kg  
4. yield_ton_per_hectare  
Write 2–3 observations from each histogram.

````python
import matplotlib.pyplot as plt

columns = [
    'rainfall_mm',
    'temperature_c',
    'fertilizer_kg',
    'yield_ton_per_hectare'
]

for col in columns:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
````

Q12. Feature Selection

Separate:

1. Input features (X)  
2. Target variable (y)  

Specify which column is being used as the target variable.

**Answer:-** yield_ton_per_hectare (Target variable)

````python
X = encoded_df.drop(
    'yield_ton_per_hectare',
    axis=1
)

y = encoded_df[
    'yield_ton_per_hectare'
]

print("Features:")
print(X.columns)

print("\nTarget:")
print(y.name)
````

**Part D: Machine Learning**

Q13. Train-Test Split

Split the dataset into:

-> 80% Training Data  
-> 20% Testing Data  

Display the shape of:

-> X_train  
-> X_test  
-> y_train  
-> y_test  

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
print("y_train:", y_train.shape)
print("y_test :", y_test.shape)
````

Q14. Linear Regression Model

1. Train a Linear Regression model.  
2. Display the model coefficients and intercept.  
3. Which feature has the highest positive coefficient?
**Answer:-** crop_type_Rice

````python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

print("Intercept:")
print(model.intercept_)

print("\nCoefficients:")

for feature, coef in zip(X.columns, model.coef_):
    print(feature, ":", coef)
````
